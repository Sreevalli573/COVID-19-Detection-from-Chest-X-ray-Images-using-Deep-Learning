```python
class PetSRDataset(Dataset):

    def __init__(self, image_paths):

        self.paths = image_paths

        self.to_tensor = transforms.ToTensor()

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):

        img = Image.open(
            self.paths[idx]
        ).convert("RGB")

        hr = img.resize(
            (128,128),
            Image.Resampling.LANCZOS
        )

        lr = hr.resize(
            (32,32),
            Image.Resampling.BICUBIC
        )

        hr = self.to_tensor(hr)
        lr = self.to_tensor(lr)

        return lr, hr
```

```python
train_dataset = PetSRDataset(train_paths)
val_dataset = PetSRDataset(val_paths)

train_loader = DataLoader(
    train_dataset,
    batch_size=16,
    shuffle=True,
    num_workers=2
)

val_loader = DataLoader(
    val_dataset,
    batch_size=16,
    shuffle=False,
    num_workers=2
)
```

```python
class DoubleConv(nn.Module):

    def __init__(self, in_ch, out_ch):

        super().__init__()

        self.conv = nn.Sequential(

            nn.Conv2d(
                in_ch,
                out_ch,
                3,
                padding=1
            ),

            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),

            nn.Conv2d(
                out_ch,
                out_ch,
                3,
                padding=1
            ),

            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):

        return self.conv(x)
```

```python
class TransformerBottleneck(nn.Module):

    def __init__(
        self,
        embed_dim=512,
        num_heads=8,
        num_layers=4
    ):

        super().__init__()

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            batch_first=True
        )

        self.transformer = nn.TransformerEncoder(
            encoder_layer,
            num_layers=num_layers
        )

    def forward(self, x):

        B,C,H,W = x.shape

        x = x.flatten(2)
        x = x.transpose(1,2)

        x = self.transformer(x)

        x = x.transpose(1,2)
        x = x.view(B,C,H,W)

        return x
```

```python
class UNetTransformerSR(nn.Module):

    def __init__(self):

        super().__init__()

        self.pool = nn.MaxPool2d(2)

        self.enc1 = DoubleConv(3,64)
        self.enc2 = DoubleConv(64,128)
        self.enc3 = DoubleConv(128,256)
        self.enc4 = DoubleConv(256,512)

        self.transformer = TransformerBottleneck(
            embed_dim=512,
            num_heads=8,
            num_layers=4
        )

        self.up1 = nn.ConvTranspose2d(
            512,
            256,
            2,
            stride=2
        )

        self.dec1 = DoubleConv(
            512,
            256
        )

        self.up2 = nn.ConvTranspose2d(
            256,
            128,
            2,
            stride=2
        )

        self.dec2 = DoubleConv(
            256,
            128
        )

        self.up3 = nn.ConvTranspose2d(
            128,
            64,
            2,
            stride=2
        )

        self.dec3 = DoubleConv(
            128,
            64
        )

        self.sr = nn.Sequential(

            nn.Conv2d(
                64,
                256,
                3,
                padding=1
            ),

            nn.PixelShuffle(2),

            nn.ReLU(),

            nn.Conv2d(
                64,
                256,
                3,
                padding=1
            ),

            nn.PixelShuffle(2),

            nn.ReLU(),

            nn.Conv2d(
                64,
                3,
                3,
                padding=1
            ),

            nn.Sigmoid()
        )

    def forward(self, x):

        e1 = self.enc1(x)

        e2 = self.enc2(
            self.pool(e1)
        )

        e3 = self.enc3(
            self.pool(e2)
        )

        e4 = self.enc4(
            self.pool(e3)
        )

        b = self.transformer(e4)

        d1 = self.up1(b)
        d1 = torch.cat(
            [d1,e3],
            dim=1
        )
        d1 = self.dec1(d1)

        d2 = self.up2(d1)
        d2 = torch.cat(
            [d2,e2],
            dim=1
        )
        d2 = self.dec2(d2)

        d3 = self.up3(d2)
        d3 = torch.cat(
            [d3,e1],
            dim=1
        )
        d3 = self.dec3(d3)

        out = self.sr(d3)

        return out
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "cpu"
)

model = UNetTransformerSR().to(device)

print(
    sum(
        p.numel()
        for p in model.parameters()
    )
)
```

```python
criterion = nn.MSELoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=1e-4
)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):

    model.train()

    train_loss = 0

    loop = tqdm(train_loader)

    for lr, hr in loop:

        lr = lr.to(device)
        hr = hr.to(device)

        pred = model(lr)

        loss = criterion(
            pred,
            hr
        )

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        train_loss += loss.item()

        loop.set_description(
            f"Epoch {epoch+1}"
        )

        loop.set_postfix(
            loss=loss.item()
        )

    print(
        f"Epoch {epoch+1}:",
        train_loss/len(train_loader)
    )
```

```python
model.eval()

lr, hr = val_dataset[0]

with torch.no_grad():

    sr = model(
        lr.unsqueeze(0).to(device)
    )

sr = sr.squeeze().cpu()

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(
    lr.permute(1,2,0)
)
plt.title("LR 32x32")

plt.subplot(1,3,2)
plt.imshow(
    sr.permute(1,2,0)
)
plt.title("SR 128x128")

plt.subplot(1,3,3)
plt.imshow(
    hr.permute(1,2,0)
)
plt.title("Ground Truth")

plt.show()
```
