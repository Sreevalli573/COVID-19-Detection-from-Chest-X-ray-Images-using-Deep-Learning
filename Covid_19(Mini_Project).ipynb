```python
# =========================
# 1. Import Libraries
# =========================
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNet

from sklearn.metrics import confusion_matrix, classification_report

# =========================
# 2. Dataset Paths
# =========================
train_path = "Covid19_ImageDataset/train"
test_path = "Covid19_ImageDataset/test"

IMG_SIZE = (224,224)
BATCH_SIZE = 32

# =========================
# 3. Data Preparation
# =========================

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Training Data
train_data = train_datagen.flow_from_directory(
    train_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

# Validation Data
val_data = train_datagen.flow_from_directory(
    train_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# Test Data
test_data = test_datagen.flow_from_directory(
    test_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

class_names = list(train_data.class_indices.keys())
print("Classes:", class_names)

# =========================
# 4. Custom CNN Model
# =========================

cnn_model = Sequential()

cnn_model.add(Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)))
cnn_model.add(MaxPooling2D(2,2))

cnn_model.add(Conv2D(64,(3,3),activation='relu'))
cnn_model.add(MaxPooling2D(2,2))

cnn_model.add(Conv2D(128,(3,3),activation='relu'))
cnn_model.add(MaxPooling2D(2,2))

cnn_model.add(Flatten())
cnn_model.add(Dense(128,activation='relu'))
cnn_model.add(Dropout(0.5))

cnn_model.add(Dense(3,activation='softmax'))

cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

cnn_model.summary()

# =========================
# 5. Train CNN Model
# =========================

cnn_history = cnn_model.fit(
    train_data,
    epochs=15,
    validation_data=val_data
)

# =========================
# 6. Transfer Learning Model
# =========================

base_model = MobileNet(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

# Freeze pretrained layers
for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128,activation='relu')(x)
x = Dropout(0.5)(x)

output = Dense(3,activation='softmax')(x)

transfer_model = Model(inputs=base_model.input, outputs=output)

transfer_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

transfer_model.summary()

# =========================
# 7. Train Transfer Model
# =========================

transfer_history = transfer_model.fit(
    train_data,
    epochs=10,
    validation_data=val_data
)

# =========================
# 8. Evaluate Models
# =========================

cnn_loss, cnn_acc = cnn_model.evaluate(test_data)
print("CNN Test Accuracy:", cnn_acc)

tl_loss, tl_acc = transfer_model.evaluate(test_data)
print("Transfer Learning Test Accuracy:", tl_acc)

# =========================
# 9. Predictions
# =========================

predictions = transfer_model.predict(test_data)
y_pred = np.argmax(predictions, axis=1)
y_true = test_data.classes

# =========================
# 10. Confusion Matrix
# =========================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=class_names,
            yticklabels=class_names)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# =========================
# 11. Classification Report
# =========================

print("Classification Report:")
print(classification_report(
    y_true,
    y_pred,
    target_names=class_names
))

# =========================
# 12. Accuracy Graph
# =========================

plt.plot(cnn_history.history['accuracy'])
plt.plot(cnn_history.history['val_accuracy'])

plt.title("CNN Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train","Validation"])
plt.show()

# =========================
# 13. Compare Models
# =========================

print("\nModel Comparison")
print("----------------------")
print("Custom CNN Accuracy:", cnn_acc)
print("Transfer Learning Accuracy:", tl_acc)
```

```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load trained model
model = load_model("transfer_model.h5")

# Image path
img_path = "/content/test_image(covid).jpg.jpg"

# Read image
img = cv2.imread(img_path)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize image to match model input
img = cv2.resize(img,(224,224))

# Normalize pixel values
img = img / 255.0

# Reshape for prediction
img_array = np.expand_dims(img, axis=0)

# Predict
prediction = model.predict(img_array)

# Class labels
classes = ["Covid","Normal","Viral Pneumonia"]

# Get predicted class
predicted_class = classes[np.argmax(prediction)]

# Show image
plt.imshow(img)
plt.title("Prediction: " + predicted_class)
plt.axis("off")
plt.show()

print("Prediction:", predicted_class)
```
